### About

This notebook trains a Siamese network of variable size on image pairs from the LFW dataset (or APD dataset). The notebook is divided roughly into the following sections:

<hr>

*Data* <br>
Using a custom dataloader, a batch of image pairs is loaded as specified in a CSV (`train_info.csv` and `test_info.csv`). Then a variety of transforms are performed on the images (*random horizontal flip, color jitter, reflection padding, random rotation up to 45 degrees, center cropping, converting to a tensor, and random erasing (scale: 0.02 to 0.10)*).

*Networks* <br>
A Siamese network modeled after a trimmed down version of resnet is defined. The overall structure of the body architecture is a customizable series of `[conv, batchnorm, relu]` blocks, followed by an end of `[batchnorm, linear, relu, linear]` layers. 

*Loss Functions* <br>
Afterward, a series of loss functions are defined (*softmax* or *cross-entropy loss*, *contrastive loss*, and *large margin cosine loss* as used in *cosface*).

*Train Eval Methods* <br>
Next, the training loop is defined, along with a validation loop. The training loop is customized such that it can optionally run a mini-validation loop within the training loop, after a specified number of batches. Additionally, a simplified validation loop is provided, which will produce a ROC curve when run (using sci-kit learn). Note that the training and validation loops were designed to accomodate triplet loss (i.e., triplets of images instead of pairs of images), as well as image pairs.

*Training Parameters*<br>
A learning rate scheduler is provided, which allows for cyclic learning rates. Other training parameters are then defined (such as an Adam optimizer, initial learning rate, weight decay, learning rate schedule, and model size).

*Train Loop*<br>
A training loop is run, with results displayed below.

*Misc* <br>
At the very end, there is a section to check the model output, as well as additional stuff present from the course of debugging.

#### SiameseDataset

```python
class SiameseDataset(data.Dataset):
    """
    simple dataset with pairs of images
    
    root: path to images
    info: dataframe containing image_1 filenames, image_2 filenames, and labels (True or False)
    transform: list of transforms, performed after opening the image
    target_transform: list of transforms for labels
    """
    def __init__(self, root, info, transform=[], label_transform=[], *args, **kwargs):
        self.root = root
        self.info = info
        
        transform = [transforms.ToTensor()] if not transform else transform
        
        self.transform = [lambda file: Image.open(self.root/file),
                          *transform
                         ]
        
        self.label_transform = label_transform
        
    def __len__(self):
        return len(self.info)
    
    def __getitem__(self, idx):
#         img1, img2, label = self.info.iloc[idx,:]
        data = self.info.iloc[idx,:-1]
        targets = self.info.iloc[idx,-1]
        
        imgs = []
        for img in data:
            for t in self.transform:
                img = t(img)
            imgs.append(img)
            
        if isinstance(targets, Iterable):
            labels = []
            for label in targets:
                for t in self.label_transform:
                    label = t(label)
                labels.append(label)
            return imgs, labels
        
        else:
            label = targets
            for t in self.label_transform:
                label = t(label)
            return imgs, label
```

#### Transforms

```python
# Transformations performed on data
trans = [transforms.RandomHorizontalFlip(),
         transforms.ColorJitter(brightness=0, contrast=(0.7,1.), saturation=(0.7,1.), hue=(0.,0.1)),
         lambda a: transforms.functional.pad(a, 120, padding_mode='reflect'),
         transforms.RandomRotation(45),
         transforms.CenterCrop(256),
         transforms.ToTensor(),
#          transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), # mean and std taken from imagenet defaults
         transforms.RandomErasing(p=0.5, scale=(0.02,0.10), value=128)
        ]

# Transformations for labels
label_trans = [
    lambda a: float(a)
]
```

##### TinySiamese

```python

# TODO: check that  res[0] has no `requires_grad=True`

class TinySiamese(nn.Module):
    def __init__(self, duplicate_factor=0, dropout=0, mom=0.1):
        """
        Siamese network with variable number of [conv, batchnorm, relu] blocks
        and an end of [batchnorm, linear, relu, linear]
        
        duplicate_factor: adds additional [conv,batchnorm,relu] blocks if (>0)
        dropout: allows for dropout to be added to network
        mom: allows for momentum to be added to batchnorm layers
        """
        
        super().__init__()
        
        self.body = Sequential(
            Conv2d(3, 64, kernel_size=(3, 3), stride=(2, 2), padding=(3, 3), bias=False),
            BatchNorm2d(64, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            ReLU(inplace=True),
            MaxPool2d(kernel_size=3, stride=2, padding=1, dilation=1, ceil_mode=False),

            *[Conv2d(64, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            BatchNorm2d(64, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            ReLU(inplace=True)]*duplicate_factor,

            # combined pooling
#             Lambda(lambda x: torch.cat([AdaptiveMaxPool2d(1)(x), AdaptiveAvgPool2d(1)(x)], 1)),
            AdaptiveAvgPool2d(1),
            
             # flatten all but the batch
#             Lambda(lambda x: x.view(x.size(0),-1)),
            Flatten(),
            
            BatchNorm1d(64, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            Dropout(p=dropout/2, inplace=False),
            Linear(in_features=64, out_features=64, bias=True),
            ReLU(inplace=True),
            
            BatchNorm1d(64, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            Dropout(p=dropout, inplace=False),
            Linear(in_features=64, out_features=64, bias=True), # out = 698
            ReLU(inplace=True)
        )

        self.tail = Sequential(
            Linear(64, 2),
#             Softmax(1)
        )
        
        self.num_layers = len(self.body)
        
        
    def forward(self, data):
        res = []
        with torch.no_grad(): res.append(self.body(data[0]))
        res.append(self.body(data[1]))
        res = torch.abs(res[1] - res[0])
        res = self.tail(res)
        return res   
```

##### SmallSiamese

```python

class SmallSiamese(nn.Module):
    def __init__(self, duplicate_factor=0, dropout=0, mom=0.1):
        """
        Siamese network with variable number of [conv, batchnorm, relu] blocks
        and an end of [batchnorm, linear, relu, linear]
        
        duplicate_factor: adds additional [conv,batchnorm,relu] blocks if (>0)
        dropout: allows for dropout to be added to network
        mom: allows for momentum to be added to batchnorm layers
        """
        
        super().__init__()
        
        self.body = Sequential(
            Conv2d(3, 64, kernel_size=(3, 3), stride=(2, 2), padding=(3, 3), bias=False),
            BatchNorm2d(64, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            ReLU(inplace=True),
            MaxPool2d(kernel_size=3, stride=2, padding=1, dilation=1, ceil_mode=False),

            *[Conv2d(64, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            BatchNorm2d(64, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            ReLU(inplace=True)]*duplicate_factor,

            Conv2d(64, 128, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1), bias=False),
            BatchNorm2d(128, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            ReLU(inplace=True),

            *[Conv2d(128, 128, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            BatchNorm2d(128, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            ReLU(inplace=True)]*duplicate_factor,

            # combined pooling
#             Lambda(lambda x: torch.cat([AdaptiveMaxPool2d(1)(x), AdaptiveAvgPool2d(1)(x)], 1)),
            AdaptiveAvgPool2d(1),
            
             # flatten all but the batch
#             Lambda(lambda x: x.view(x.size(0),-1)),
            Flatten(),
            
            BatchNorm1d(256, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            Dropout(p=dropout/2, inplace=False),
            Linear(in_features=256, out_features=128, bias=True),
            ReLU(inplace=True),
            
            BatchNorm1d(128, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            Dropout(p=dropout, inplace=False),
            Linear(in_features=128, out_features=128, bias=True), # out = 698
            ReLU(inplace=True)
        )
        
        self.tail = Linear(128, 2)
        
        self.num_layers = len(self.body)
        
        
    def forward(self, data):
        res = []
        for i in range(len(data)): # Siamese nets; sharing weights
            x = data[i]
            x = self.body(x)

            res.append(x)

        res = torch.abs(res[1] - res[0])
        res = self.tail(res)
        return res
```

##### MedSiamese

```python

class MedSiamese(nn.Module):
    def __init__(self, duplicate_factor=0, dropout=0):
        """
        Siamese network with variable number of [conv, batchnorm, relu] blocks
        and an end of [batchnorm, linear, relu, linear]
        
        duplicate_factor: adds additional [conv,batchnorm,relu] blocks if (>0)
        dropout: allows for dropout to be added to network
        """
        
        super().__init__()
        
        self.body = Sequential(
            Conv2d(3, 64, kernel_size=(3, 3), stride=(2, 2), padding=(3, 3), bias=False),
            BatchNorm2d(64, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            ReLU(inplace=True),
            MaxPool2d(kernel_size=3, stride=2, padding=1, dilation=1, ceil_mode=False),

            *[Conv2d(64, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            BatchNorm2d(64, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            ReLU(inplace=True)]*duplicate_factor,

            Conv2d(64, 128, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1), bias=False),
            BatchNorm2d(128, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            ReLU(inplace=True),

            *[Conv2d(128, 128, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            BatchNorm2d(128, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            ReLU(inplace=True)]*duplicate_factor,

            Conv2d(128, 256, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1), bias=False),
            BatchNorm2d(256, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            ReLU(inplace=True),

            Conv2d(256, 256, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            BatchNorm2d(256, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),

            *[nn.Conv2d(256, 256, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(256, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(inplace=True)]*duplicate_factor,
            
            # combined pooling
#             Lambda(lambda x: torch.cat([AdaptiveMaxPool2d(1)(x), AdaptiveAvgPool2d(1)(x)], 1)),
            AdaptiveAvgPool2d(1),
            
             # flatten all but the batch
#             Lambda(lambda x: x.view(x.size(0),-1)),
            Flatten(),

            BatchNorm1d(612, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            Dropout(p=dropout/2, inplace=False),
            Linear(in_features=612, out_features=256, bias=True),
            ReLU(inplace=True),
            
            BatchNorm1d(256, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            Dropout(p=dropout, inplace=False),
            Linear(in_features=256, out_features=256, bias=True), # out = 698
            ReLU(inplace=True)
        )
        self.tail = Linear(256, 2)
        
        self.num_layers = len(self.body)
        
        
    def forward(self, data):
        res = []
        for i in range(2): # Siamese nets; sharing weights
            x = data[i]
            x = self.body(x)
            res.append(x)

        res = torch.abs(res[1] - res[0])
        res = self.tail(res)
        return res
```

##### Siamese

```python

class Siamese(nn.Module):
    def __init__(self, duplicate_factor=0, dropout=0):
        """
        Siamese network with variable number of [conv, batchnorm, relu] blocks
        and an end of [batchnorm, linear, relu, linear]
        
        duplicate_factor: adds additional [conv,batchnorm,relu] blocks if (>0)
        dropout: allows for dropout to be added to network
        """
        super().__init__()
        
        self.body = Sequential(
            Conv2d(3, 64, kernel_size=(3, 3), stride=(2, 2), padding=(3, 3), bias=False), # originally 7x7
            BatchNorm2d(64, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            ReLU(inplace=True),
            MaxPool2d(kernel_size=3, stride=2, padding=1, dilation=1, ceil_mode=False),

            *[Conv2d(64, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            BatchNorm2d(64, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            ReLU(inplace=True)]*duplicate_factor,

            Conv2d(64, 128, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1), bias=False),
            BatchNorm2d(128, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            ReLU(inplace=True),

            *[Conv2d(128, 128, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            BatchNorm2d(128, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            ReLU(inplace=True)]*duplicate_factor,

            Conv2d(128, 256, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1), bias=False),
            BatchNorm2d(256, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            ReLU(inplace=True),

            Conv2d(256, 256, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            BatchNorm2d(256, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),

            *[nn.Conv2d(256, 256, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(256, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(inplace=True)]*duplicate_factor,

            Conv2d(256, 512, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1), bias=False),
            BatchNorm2d(512, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            ReLU(inplace=True),

            *[nn.Conv2d(512, 512, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(512, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(inplace=True)]*duplicate_factor,
            
            # combined pooling
#             Lambda(lambda x: torch.cat([AdaptiveMaxPool2d(1)(x), AdaptiveAvgPool2d(1)(x)], 1)),
            AdaptiveAvgPool2d(1),
            
             # flatten all but the batch
#             Lambda(lambda x: x.view(x.size(0),-1)),
            Flatten(),
            
            BatchNorm1d(512, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            Dropout(p=dropout/2, inplace=False),
            Linear(in_features=512, out_features=512, bias=True),
            ReLU(inplace=True),
            
            BatchNorm1d(512, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            Dropout(p=dropout, inplace=False),
            Linear(in_features=512, out_features=512, bias=True), # out = 698
            ReLU(inplace=True)
        )
        self.tail = Linear(512, 2)
        
        self.num_layers = len(self.body)
        
        
    def forward(self, data):
        res = []
        for i in range(2): # Siamese nets; sharing weights
            x = data[i]
            x = self.body(x)
            res.append(x)

        res = torch.abs(res[1] - res[0])
        res = self.tail(res)
        return res
```

Contrastive Loss

$$ L = \dfrac{1}{2}(1-Y)(D_W^2) + \dfrac{1}{2} Y (\mathrm{max}(0, m-D_W))^2 $$

```python
class ContrastiveLoss(nn.Module):
    """
    L(X_1, X_2, Y) = (1-Y)*0.5*(Dw**2) + (Y)*0.5*(max(0, m - Dw))**2
    
    m: margin (>0), used to create radius; disimilar pairs only contribute to loss if within margin
    Dw: euclidean distance function on outputs of X_1 and X_2
    Y: binary target label
    W: weights
    
    Reference: H. Hadsell, S. Chopra, Y. LeCun. Dimensionality Reduction by Learning an Invariant Mapping
    http://yann.lecun.com/exdb/publis/pdf/hadsell-chopra-lecun-06.pdf
    
    Additional reference:
    https://github.com/adambielski/siamese-triplet/blob/master/losses.py
    """
    def __init__(self, m):
        super(ContrastiveLoss, self).__init__()
        self.m = m  # set margin
        
    def forward(self, outputs, Y):
        output1 = outputs[:,0]
        output2 = outputs[:,1]
        Y = Y.float()
        Dw = (output2 - output1).pow(2) # sqrt() only performed on on Dw below for stability
        L = 0.5*(1-Y)*(Dw) + 0.5*Y*(torch.relu(self.m - Dw.sqrt())).pow(2)
        return torch.mean(L)
```

Large Margin Cosine Loss

$$L = \dfrac{1}{N} \sum -log \dfrac{e^{s\cdot (cos\theta_{y_i,i} - m)}}{e^{s\cdot (cos\theta_{y_i,i} - m)} + \sum_{j\neq y_i} e^{s\cdot cos\theta_{y_j}}} $$

Where $W$ and $x$ are normalized, and:

- $cos(\theta_j,i) = W_j^Tx_i$
- 0 < m < 45
- s was set to 64 empirically

```python
class LargeMarginCosineLoss(nn.Module):
    """
    Reference: 
    H. Wang et al. CosFace: Large Margin Cosine Loss for Deep Face Recognition
    https://arxiv.org/pdf/1801.09414.pdf
    
    Also referenced cvqluu's implementation of Angular Penalty Loss:
    https://paperswithcode.com/paper/cosface-large-margin-cosine-loss-for-deep
    """
    
    def __init__(self, in_features=2, out_features=2, s=64.0, m=0.35):
        super(LargeMarginCosineLoss, self).__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.s = s
        self.m = m
        
        # cos(θ_j,i) = W_j^T * x_i
        self.linear = Linear(in_features, out_features, bias=False)
        
    def forward(self, x, targets):
        
        # normalize
        x = F.normalize(x, p=2, dim=1)
        for W in self.linear.parameters():
            W = F.normalize(W, p=2, dim=1)
            
        cos_θ = self.linear(x)
        s_cos_θ = self.s*torch.diagonal(cos_θ.transpose(0,1)[targets]-self.m)
        
        cos_θj = [torch.cat((cos_θ[j,:y], cos_θ[j,(y+1):])).unsqueeze(0) for j, y in enumerate(targets)]
        sum_j = torch.sum(torch.exp(self.s*torch.cat(cos_θj, dim=0)), dim=1)
        
        result = torch.mean(torch.log(torch.exp(s_cos_θ) + sum_j) - torch.log(torch.exp(s_cos_θ)))
        
        return result
```

#### Train Eval Method

```python
def train_eval(model, dataloader, valid_dataloader, opt, epoch, loss_func=F.cross_entropy, \
               device=None, verbose=True, disp=None, lr_sched=None, \
               mini_epoch_length = 20, run_val_loop=True, full_epoch=False,\
               roc_curve=None, roc_auc=None):
    """
    Training loop, with an optional validation loop embedded inside it
    
    dataloader: training dataloader
    valid_dataloader: validation dataloader (used in validation loop)
    opt: optimizer
    epoch: current epoch number
    loss_func: loss function
    lr_sched: torch.optim.lr_scheduler, for scheduled learning rates if available
    
    device: torch.device; if None, checks if gpu is available and sets defaults
    
    verbose: if True, displays updated information
    disp: results to be displayed, in pandas.DataFrame() format
    
    mini_epoch_length: number of batches done prior to validation; default is 20 to shorten training time
    run_val_loop: True if you want to run a validation loop inside of train_eval;
                if False, just runs training loop
    full_epoch: if True, then each epoch loops over all training data; 
                otherwise, each epoch is mini_epoch_length
    
    """
    timer = time.time()
    model.train()
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu') if (device==None) else device

    for iteration, (xb, yb) in enumerate(dataloader):
        # xb: all pairs of images in a batch (if batch_size=64, then data contains 64 pairs of images)
        # yb: labels for entire batch
        
        # move to device
        yb = yb.type(torch.LongTensor).to(device)
        for i in range(len(xb)):
            xb[i] = xb[i].to(device)
        
        # check number of images
        triplet = True if (len(xb) > 2) else False

        opt.zero_grad()
        
        if not triplet:
            output = model(xb[:2])
            loss = loss_func(output, yb)
            loss.backward()
            opt.step()
            
        else:
            output_A = model([xb[0], xb[1]]) # xb[:2]
            output_B = model([xb[0], xb[2]]) # xb[0:3:2]

            yb_A = torch.squeeze(yb[:,0])
            yb_B = torch.squeeze(yb[:,1])

            loss = loss_func(output_A, yb_A) + loss_func(output_B, yb_B)
#             loss = loss_func(output_A, output_B, yb_A)
            loss.backward()
            opt.step()
            
        timer2 = time.time() - timer
        
        # use scheduled learning rate after each batch, if scheduler is available
        if (lr_sched != None): lr_sched.step()
        
        if isinstance(disp, pd.core.frame.DataFrame):
            disp.loc[epoch, "Time"] = str(datetime.timedelta(seconds=round(timer2)))
            disp.loc[epoch, "Train_Loss"] = loss.item()
            disp.loc[epoch, "Epoch"] = epoch
    
        
        sample_num = iteration*dataloader.batch_size
        total = len(dataloader.dataset)
        pct_progress = 100.*(float(sample_num)/float(total))
        
        if verbose:    
            clear_output(wait=True)
            display(f'({iteration}) {sample_num}/{total} ({pct_progress:.0f}%)  Loss: {loss.item():.6f}')
            display(HTML(disp.to_html(index=False)))
            
        if (run_val_loop) and (iteration % mini_epoch_length == 0) and (iteration > 0):
            # run a validation loop
            validation_results = validate(model, valid_dataloader, 
                                          sample_num=sample_num, train_tot=total, epoch=epoch, 
                                          loss_func=loss_func, device=device, disp=disp, 
                                          verbose=verbose)
            
            disp, roc_curve, roc_auc = validation_results
            
            
            if not full_epoch:
                epoch += 1
                timer = time.time() # reset timer
                disp.loc[len(disp)] = [0,0,"","",0,0,0,0,0,0,""] # add a new row to the display dataframe
    
    if full_epoch:
        disp.loc[len(disp)] = [0,0,"","",0,0,0,0,0,0,""] # add a new row to the display dataframe    
    
    return disp, roc_curve, roc_auc, epoch
    
            
```

#### Validate Method

```python
def validate(model, dataloader, epoch, sample_num,\
             train_tot, loss_func=F.cross_entropy, \
             device=None, disp=None, val_epoch_length=15, verbose=True):
    """
    Validation loop that produces ROC curve, currently stored inside train_eval()
    
    dataloader: validation dataloader
    epoch: number of epochs
    device: torch.device; if None, checks if gpu is available and sets defaults
    loss_func: loss function
    verbose: if True, displays updated information
    disp: results to be displayed, in pandas.DataFrame() format
    val_epoch_length: number of validation batches done in an epoch; default is 15 to shorten validation time
    
    sample_num: used for display; current number of samples trained
    train_tot: used for display; current total number of samples trained
    """
    model.eval()
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu') if not device else device
    
    num_correct = 0
    loss = 0
    tot = 0
    
#     timer = time.time()
    
    with torch.no_grad():
        
        TP = 0
        FP = 0
        TN = 0
        FN = 0
        
        for iteration, (xb, yb) in enumerate(dataloader):
            # xb: all pairs of images in a batch (if batch_size=64, then data contains 64 pairs of images)
            # yb: labels for entire batch
            
            # move to device
            yb = yb.type(torch.LongTensor).to(device)
            for i in range(len(xb)):
                xb[i] = xb[i].to(device)
                
            
            # check number of images
            triplet = True if (len(xb) > 2) else False
            
            if not triplet:
                output = model(xb[:2])
                loss = loss_func(output, yb)
                preds_xb = torch.argmax(output, dim=1)
                
                num_correct += torch.sum(preds_xb == yb).cpu()
                tot += len(yb)
                acc = 100.*(float(num_correct) / float(tot))
                
                TP += torch.sum((preds_xb == yb) & (preds_xb == 1)).cpu().item()
                FP += torch.sum((preds_xb != yb) & (preds_xb == 1)).cpu().item()
                TN += torch.sum((preds_xb == yb) & (preds_xb == 0)).cpu().item()
                FN += torch.sum((preds_xb != yb) & (preds_xb == 0)).cpu().item()
                Total = np.sum([TP,FP,TN,FN])
                
                roc_curve, roc_auc = generate_ROC(preds_xb, yb)
                
            else:
                output_A = model(xb[:2])
                output_B = model(xb[0:3:2])

                yb_A = torch.squeeze(yb[:,0])
                yb_B = torch.squeeze(yb[:,1])
                
                loss = loss_func(output_A, yb_A) + loss_func(output_B, yb_B)
                
                preds_xb_A = torch.argmax(output_A, dim=1)
                preds_xb_B = torch.argmax(output_B, dim=1)
                
                correct_A = torch.sum(preds_xb_A == yb_A).cpu()
                correct_B = torch.sum(preds_xb_B == yb_B).cpu()
                
                num_correct += correct_A + correct_B
                tot += len(yb_A) + len(yb_B)
                acc = 100.*(float(num_correct) / float(tot))
                
                TP_A = torch.sum((preds_xb_A == yb) & (preds_xb_A == 1)).cpu().item()
                FP_A = torch.sum((preds_xb_A != yb) & (preds_xb_A == 1)).cpu().item()
                TN_A = torch.sum((preds_xb_A == yb) & (preds_xb_A == 0)).cpu().item()
                FN_A = torch.sum((preds_xb_A != yb) & (preds_xb_A == 0)).cpu().item()
                
                TP_B = torch.sum((preds_xb_B == yb) & (preds_xb_B == 1)).cpu().item()
                FP_B = torch.sum((preds_xb_B != yb) & (preds_xb_B == 1)).cpu().item()
                TN_B = torch.sum((preds_xb_B == yb) & (preds_xb_B == 0)).cpu().item()
                FN_B = torch.sum((preds_xb_B != yb) & (preds_xb_B == 0)).cpu().item()
                
                TP += TP_A + TP_B
                FP += FP_A + FP_B
                TN += TN_A + TN_B
                FN += FN_A + FN_B
                Total = np.sum([TP,FP,TN,FN])
            
                roc_curve, roc_auc = generate_ROC_mult([preds_xb_A, preds_xb_B], yb)
            
#             timer2 = time.time() - timer
            
            if isinstance(disp, pd.core.frame.DataFrame):
                disp.loc[epoch, "Valid_Loss"] = float(loss)
                disp.loc[epoch, "Valid_Accuracy"] = float(acc)
                disp.loc[epoch, "TP"] = float(TP)
                disp.loc[epoch, "FP"] = float(FP)
                disp.loc[epoch, "TN"] = float(TN)
                disp.loc[epoch, "FN"] = float(FN)
                disp.loc[epoch, "Total"] = float(Total)
                disp.loc[epoch, "ROC_AUC"] = roc_auc
                
                
#                 disp.loc[epoch, "Time"] = str(datetime.timedelta(seconds=round(timer2)))
            
            if verbose:
                clear_output(wait=True)
                display(f'({iteration}) {sample_num}/{train_tot} Loss: {loss:.6f}, Acc.: {acc:.4f}% ({num_correct}/{tot})')
                display(HTML(disp.to_html(index=False)))
            
            if (iteration % val_epoch_length == 0) and (iteration > 0):
                break
            
    return disp, roc_curve, roc_auc
```

##### generate_ROC

```python

def generate_ROC(predictions, true_values, show_plt=False):
    """
    Generates an ROC curve, given true values and predicted values
    """

    score = np.array([predictions[i] for i in range(predictions.shape[0])])
    fpr, tpr, thresholds = metrics.roc_curve(true_values, score, pos_label=1)
    roc_auc = metrics.auc(fpr, tpr)
    
    # Plot ROC Curve
    lw = 2
    fig, ax = plt.subplots(1,1)
    ax.plot(fpr,tpr,color='navy',
             lw=lw, label=f'ROC curve (area = {roc_auc: 0.2f})')
    ax.plot([0,1],[0,1], color='navy', lw=lw, linestyle='--',alpha=0.15)
    ax.set(xlabel='(1–Specificity) - False Positive Rate',
          ylabel='Sensitivity - True Positive Rate',
          title='Receiver Operating Characteristic')
    ax.legend(loc="lower right")
    ax.grid()
    
    if show_plt: 
        fig.show()
    plt.close();

    return fig, roc_auc
```

```python

def generate_ROC_mult(prediction_set, true_values, show_plt=False):
    """
    Generates an ROC curve for multiple sets of images (triplets), given true values and predicted values
    prediction_set: sets of predictions for each image (for multiple images)
    """
    lw = 2
    fig, ax = plt.subplots(1,1)
    ax.plot([0,1],[0,1], color='navy', lw=lw, linestyle='--',alpha=0.15)
    
    for (i, predictions) in enumerate(prediction_set):
        score = np.array([predictions[i] for i in range(predictions.shape[0])])
        fpr, tpr, thresholds = metrics.roc_curve(true_values, score, pos_label=1)
        roc_auc = metrics.auc(fpr, tpr)

        # Plot ROC Curve for Each Image Set
        lw = 2
        fig, ax = plt.subplots(1,1)
        ax.plot(fpr,tpr,
                 lw=lw, label=f'ROC curve {i} (area = {roc_auc: 0.2f})')
    ax.set(xlabel='(1–Specificity) - False Positive Rate',
          ylabel='Sensitivity - True Positive Rate',
          title='Receiver Operating Characteristic')
    ax.legend(loc="lower right")
    ax.grid()
    
    if show_plt: 
        fig.show()
    plt.close();

    return fig, roc_auc
    
    
```

#### LR Scheduler

The learning rate is scheduled with the following function to use cyclic learning rates

```python
# LR sched function
def sched_lr(min_pct, max_pct, cycle_length, iterations):
    """
    Formula for a cyclic learning rate; returns an array of learning rates
    
    min_pct: minimum percentage of initial learning rate
    max_pct: maximum percentage of initial learning rate
    cycle: steps before each cycle
    iterations: number of iterations (length of array)
    """
    lrs = [max(min_pct, max_pct*abs(math.cos(3*(i%cycle_length)/(2*cycle_length)))) \
           for i in range(iterations)]
    return lrs

xs = sched_lr(0.5, 1, 20, 40)

print(len(xs))
plt.plot(xs)
```

```python

class TinySiamese(nn.Module):
    def __init__(self, duplicate_factor=0, dropout=0, mom=0.1):
        super().__init__()
        
        self.body = Sequential(
            Conv2d(3, 64, kernel_size=(3, 3), stride=(2, 2), padding=(3, 3), bias=False),
            BatchNorm2d(64, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            ReLU(inplace=True),
            MaxPool2d(kernel_size=3, stride=2, padding=1, dilation=1, ceil_mode=False),

            *[Conv2d(64, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            BatchNorm2d(64, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            ReLU(inplace=True)]*duplicate_factor,

            # combined pooling
            Lambda(lambda x: torch.cat([AdaptiveMaxPool2d(1)(x), AdaptiveAvgPool2d(1)(x)], 1)),

             # flatten all but the batch
            Lambda(lambda x: x.view(x.size(0),-1)),
            
            BatchNorm1d(128, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            Dropout(p=dropout/2, inplace=False),
            Linear(in_features=128, out_features=64, bias=True),
            ReLU(inplace=True),
            
            BatchNorm1d(64, eps=1e-05, momentum=mom, affine=True, track_running_stats=True),
            Dropout(p=dropout, inplace=False),
            Linear(in_features=64, out_features=64, bias=True), # out = 698
            ReLU(inplace=True)
        )
        
        self.tail = Linear(128, 2)
        
        self.num_layers = len(self.body)
        
        
    def forward(self, data):
        res = []
        for i in range(len(data)): # Siamese nets; sharing weights
            x = data[i]
            x = self.body(x)

            
            res.append(x)

        res = torch.abs(res[1] - res[0])
        res = self.tail(res)
        return res    
```

```python
"""
NOTE: This was not used in my most recently used models

Generic Lambda layers

Source: https://discuss.pytorch.org/t/could-anyone-explain-how-a-specific-layer-works-in-my-model/14085

"""

class LambdaBase(nn.Sequential):
    def __init__(self, fn, *args):
        super(LambdaBase, self).__init__(*args)
        self.lambda_func = fn

    def forward_prepare(self, input):
        output = []
        for module in self._modules.values():
            output.append(module(input))
        return output if output else input

class Lambda(LambdaBase):
    def forward(self, input):
        return self.lambda_func(self.forward_prepare(input))
```
